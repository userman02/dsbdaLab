```python
import pandas as pd

# Load the Air Quality datasets
air_quality_data = pd.read_csv('Downloads\\Air-Quality.csv')
heart_diseas_data = pd.read_csv('Downloads\\Heart.csv')

# Display the first few rows of each dataset
print("Air Quality Dataset :")
print(air_quality_data.head())

```

```python
print("Heart Dataset :")
print(heart_diseas_data.head())
```

```python
import matplotlib.pyplot as plt

# Plot histogram of AQI
plt.hist(air_quality_data['Wind'], bins=20)
plt.title('Distribution of Air-Wind ')
plt.xlabel('Wind')
plt.ylabel('Temp')
plt.show()
```

```python
import seaborn as sns

# Plot boxplot of heart rate by heart disease diagnosis
sns.boxplot(data=heart_diseas_data, x='RestBP', y='MaxHR')
plt.title('Heart Rate by RestBp Diagnosis')
plt.xlabel('RestBP')
plt.ylabel('Heart Rate')
plt.show()
```

```python
import seaborn as sns

# Plot scatterplot of NO2 and O3 levels
sns.scatterplot(data=air_quality_data, x='Ozone', y='Solar.R')
plt.title('Ozone vs. Solar.R Levels')
plt.xlabel('Ozone')
plt.ylabel('Solar.R')
plt.show()
```

```python
import seaborn as sns

# Calculate correlation matrix
correlation_matrix = air_quality_data.corr()

# Plot heatmap of the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
