```python
import pandas as pd

# Load the forest fire dataset
forest_fire_data = pd.read_csv('C:\\Users\\acer\\Downloads\\archive (8)\\forestfires.csv')

```

```python
forest_fire_data
```

```python
import matplotlib.pyplot as plt

# Plot histogram of the forest fire area
plt.hist(forest_fire_data['area'], bins=20)
plt.title('Distribution of Forest Fire Area')
plt.xlabel('Area')
plt.ylabel('Frequency')
plt.show()
```

```python
import seaborn as sns

# Plot boxplot of the forest fire area by month
sns.boxplot(data=forest_fire_data, x='month', y='area')
plt.title('Forest Fire Area by Month')
plt.xlabel('Month')
plt.ylabel('Area')
plt.show()
```

```python
import seaborn as sns

# Plot scatterplot of temperature vs. relative humidity
sns.scatterplot(data=forest_fire_data, x='temp', y='RH')
plt.title('Temperature vs. Relative Humidity')
plt.xlabel('Temperature')
plt.ylabel('Relative Humidity')
plt.show()
```

```python
import seaborn as sns

# Calculate correlation matrix
correlation_matrix = forest_fire_data.corr()

# Plot heatmap of the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
