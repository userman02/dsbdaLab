```python
import matplotlib.pyplot as plt  #2-visualize data
import seaborn as sns
```

```python
# Plot the distribution of the target variable (e.g., 'area')
plt.figure(figsize=(8, 6))
sns.histplot(data=forest_fire_data, x='area', bins=20)
plt.title('Distribution of Forest Fire Area')
plt.xlabel('Area')
plt.ylabel('Count')
plt.show()
```

```python
# Plot the correlation matrix of the dataset
plt.figure(figsize=(10, 8))
sns.heatmap(forest_fire_data.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
# data mining 1) feature selection
# Select relevant features for analysis
selected_features = ['temp', 'RH', 'wind', 'rain', 'area']
forest_fire_subset = forest_fire_data[selected_features]
```

```python
forest_fire_subset
```

```python
#clustering
from sklearn.cluster import KMeans

# Perform clustering using K-means algorithm
kmeans = KMeans(n_clusters=3, random_state=42)
forest_fire_clustered = forest_fire_subset.copy()
forest_fire_clustered['cluster'] = kmeans.fit_predict(forest_fire_subset)


```

```python
forest_fire_clustered['cluster'] 
```

```python
# Print the cluster centers
print(kmeans.cluster_centers_)
```

```python
# Visualize the clusters
sns.scatterplot(data=forest_fire_clustered, x='temp', y='RH', hue='cluster')
plt.title('Clusters of Forest Fire Data')
plt.xlabel('Temperature')
plt.ylabel('Relative Humidity')
plt.show()
```
